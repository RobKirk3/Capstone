# Next let's get the Audio Features

```python
# this code from https://nvbn.github.io/2019/10/14/playlist-analysis/
features = []
for n, chunk_series in full_track_info.groupby(np.arange(len(full_track_info)) // 50).id:
    features += sp.audio_features([*map(str, chunk_series)])
features_df = pd.DataFrame.from_dict(filter(None, features))
# and here we are merging it with our previous dataframe
tracks_with_features_df = full_track_info.merge(features_df, on=['id'], how='inner')
```
